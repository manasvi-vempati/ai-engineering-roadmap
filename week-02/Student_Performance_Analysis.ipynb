```python
# ==========================================
# Student Performance Dataset Analysis
# ==========================================

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ---------------------------
# 1. Load Dataset
# ---------------------------

df = pd.read_csv('Performance.csv')

# Display first 5 rows
print("First 20 Rows:")
display(df.head(20))
```

```python
# -----------------------------
# 2. Dataset Overview
# -----------------------------

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
# -------------------------------
# 3. Missing Value Check
# -------------------------------

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# ------------------------------
# 4. Duplicate Check
# ------------------------------

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
# --------------------------------
# 5. Column Type Inspection
# --------------------------------

print("\nData Types:")
print(df.dtypes)
```

```python
# ------------------------------
# 6. Statistical Summary
# ------------------------------

print("\n" + "=" * 60)
print("📈 DESCRIPTIVE STATISTICS")
print("=" * 60)

desc_stats = df.describe()
display(desc_stats.round(2))

# Key Insights
print("\n KEY INSIGHTS:")
print(f"• Average Final Exam Score: {df['Final Exam Marks (out of 100)'].mean():.1f}")
print(f"• Highest Score: {df['Final Exam Marks (out of 100)'].max()}")
print(f"• Lowest Score: {df['Final Exam Marks (out of 100)'].min()}")
print(f"• Average Study Hours: {df['Daily Study Hours'].mean():.1f} hours/day")
print(f"• Average Attendance: {df['Attendance (%)'].mean():.1f}%")
```

```python
# -------------------------
# 7. Sorting
# -------------------------

print("\nTop 10 Students by Final Exam Marks:")

sorted_df = df.sort_values(
    by='Final Exam Marks (out of 100)',
    ascending=False
)

display(sorted_df.head(10))
```

```python
# --------------------------
# 8. Filtering
# --------------------------

print("\nStudents Scoring Above 80:")

high_scorers = df[
    df['Final Exam Marks (out of 100)'] > 80
]

print("Number of Students:", len(high_scorers))
display(high_scorers.head(10))
```

```python
# =============================
# CHART 1 : BAR CHART
# =============================

avg_scores = df[
    [
        'Internal Test 1 (out of 40)',
        'Internal Test 2 (out of 40)',
        'Final Exam Marks (out of 100)'
    ]
].mean()

plt.figure(figsize=(8,5))
avg_scores.plot(kind='bar')

plt.title('Average Scores Comparison')
plt.xlabel('Assessment Type')
plt.ylabel('Average Score')
plt.grid(True)

plt.show()
```

```python
# ==============================
# CHART 2 : LINE CHART
# ==============================

plt.figure(figsize=(10,5))

plt.plot(
    df['Attendance (%)'].head(50),
    marker="o"
)

plt.title('Attendance Trend (First 50 Students)')
plt.xlabel('Student Index')
plt.ylabel('Attendance (%)')
plt.grid(True)

plt.show()
```

```python
# ==========================
# CHART 3 : HISTOGRAM
# ==========================

plt.figure(figsize=(10,5))

plt.hist(
    df['Final Exam Marks (out of 100)'],
    bins=15
)

plt.title('Distribution of Final Exam Marks')
plt.xlabel('Final Exam Marks')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```
